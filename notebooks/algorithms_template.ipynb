## 1) 問い
- 入力: ソート済み配列 `A` と値 `x`
- 出力: `x` の位置（無ければ -1）

## 2) 擬似コード
1. `low=0, high=n-1`
2. `while low<=high`:
3. `mid=(low+high)//2`
4. `A[mid] ? x` で分岐

````python
def binary_search(a, x):
    low, high = 0, len(a)-1
    while low <= high:
        mid = (low + high) // 2
        if a[mid] == x:
            return mid
        if a[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return -1

assert binary_search([1,2,3,4,5], 4) == 3
assert binary_search([1,2,3,4,5], 6) == -1
print('ok')
````

## 5) 簡単な可視化（色指定なし）

````python
import matplotlib.pyplot as plt
xs = list(range(1,11))
ys = [binary_search(xs, k) for k in xs]
plt.figure(figsize=(5,3))
plt.plot(xs, ys, marker='o')
plt.title('Index vs Value (Sorted Array)')
plt.xlabel('Value')
plt.ylabel('Index Found')
plt.tight_layout()
plt.show()
````

## 1. 多角的問題分析

### 競技プログラミング視点
- **制約**: `1 ≤ l, b ≤ 1000` → 小規模入力、数学的解法が最適
- **核心**: 最大の正方形サイズ = `gcd(l, b)`、個数 = 面積 / 正方形面積

### 業務開発視点
- **型安全性**: 入力は整数、GCD計算で標準ライブラリ活用
- **エッジケース**: `l = b` の場合（正方形そのもの）も正しく処理

### Python特有考慮
- `math.gcd()` は C実装で高速（O(log min(l,b))）
- 整数除算 `//` で型安全性確保

## 3. 採用アルゴリズムと根拠

### 数学的洞察
1. 正方形の一辺は `l` と `b` の両方で割り切れる必要がある
2. 最大サイズ = 最大公約数 `g = gcd(l, b)`
3. 個数 = `(l / g) × (b / g) = (l × b) / g²`

### 検証
- 例1: `l=2, b=2` → `g=2` → `(2×2)/(2×2) = 1` ✓
- 例2: `l=6, b=9` → `g=3` → `(6×9)/(3×3) = 6` ✓

## 4. HackerRankでの回答フォーマット

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'restaurant' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER l
#  2. INTEGER b
#

def restaurant(l: int, b: int) -> int:
    """
    パンを最大サイズの正方形に切り分けた際の個数を計算
    
    アルゴリズム:
    - 正方形の最大サイズ = gcd(l, b)
    - 個数 = (l * b) / gcd(l, b)^2
    
    Time Complexity: O(log(min(l, b)))
    Space Complexity: O(1)
    
    Args:
        l: パンの長さ
        b: パンの幅
    
    Returns:
        最大サイズの正方形の個数
    """
    g = math.gcd(l, b)
    return (l * b) // (g * g)


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input().strip())

    for t_itr in range(t):
        first_multiple_input = input().rstrip().split()

        l = int(first_multiple_input[0])

        b = int(first_multiple_input[1])

        result = restaurant(l, b)

        fptr.write(str(result) + '\n')

    fptr.close()
```
